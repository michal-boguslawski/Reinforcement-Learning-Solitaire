```python
import gymnasium as gym
from gymnasium.wrappers import NumpyToTorch, DtypeObservation
import numpy as np
from torch import nn
import torch as T
```

```python
env = gym.make("CarRacing-v3", continuous=True)
env = DtypeObservation(env, np.float32)
env = NumpyToTorch(env)
```

```python
env.action_space
```

```python
env.observation_space
```

```python
state, info = env.reset()
state.shape, type(state), info, state.max(), state.min()
```

```python
transposed_state = state.permute(2, 0, 1).to(T.float32)
transposed_state.shape, transposed_state.dtype
```

```python
layer1 = nn.Conv2d(
    in_channels=3,
    out_channels=64,
    kernel_size=7,
    stride=3
)
x1 = layer1(transposed_state)
x1.shape
```

```python
layer2 = nn.Conv2d(
    in_channels=64,
    out_channels=128,
    kernel_size=5,
    stride=2
)
x2 = layer2(x1)
x2.shape
```

```python
layer3 = nn.Conv2d(
    in_channels=128,
    out_channels=256,
    kernel_size=5,
    stride=2
)
x3 = layer3(x2)
x3.shape
```

```python
layer4 = nn.Conv2d(
    in_channels=256,
    out_channels=256,
    kernel_size=3,
    stride=1
)
x4 = layer4(x3)
x4.shape
```
